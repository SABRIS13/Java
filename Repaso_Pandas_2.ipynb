```python
labels = ['1 y 2 cuartos', '3 y 4 cuartos', '5 y 6 cuartos', '7 cuartos ó más']
```

#### <font color= 'purple'> Función cut

La *función cut()* es una herramienta de pandas que ayuda a crear distribuciones de frecuencia.Es posible crear labels para las clases creadas por la función cut().

La función cut() permite especificar los límites de cada clase.

```python
cuartos = pd.cut(datos.Cuartos, clases, labels = labels)
```

```python
pd.value_counts(cuartos)
```

```python
#Tomar en cuenta alquileres con 0 cuartos
cuartos = pd.cut(datos.Cuartos, clases, labels = labels, include_lowest= True) 
```

```python
pd.value_counts(cuartos)
```
